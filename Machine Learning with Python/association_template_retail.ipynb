```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori
```

## Import data

```python
df = pd.read_excel('./data/summary_transactions.xlsx', header = None)
df.head()
```

```python
df.info
```

## Pre-processing

```python
records = []
for i in range(0, df.shape[0]):
    records.append([str(df.values[i, j]) for j in range(0, df.shape[1])])
```

## Algorithm definition APRIORI
### Threshold definition

```python
min_n = 2

# Calculate support. Suppose we have data from a 7 day (week) range, and we want products bought at least 3 times a day:
min_day = 3
range_data = 7
min_support = min_day * range_data / len(df)

min_confidence = 0.2
min_lift = 3
```

### Algorithm execution

```python
association_rules = apriori(records, min_support=min_support, min_confidence=min_confidence, min_lift=min_lift, min_length=min_n)
association_rules = list(association_rules)
```

```python
len(association_rules)
```

```python
for item in association_rules:
    # first index of the inner list
    pair = item[0]
    items = [x for x in pair]
    # second index of the inner list
    support = item[1]

    print(f"Rule: {items[0]} -> {items[1]}\nSupport: {support}")
    print(f"Confidence: {item[2][0][2]}\nLift: {item[2][0][3]}")
    print(f"------------------------------")
```
