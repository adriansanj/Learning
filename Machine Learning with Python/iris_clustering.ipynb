Task: Develop a machine learning clustering model to classify iris species

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

## Import data

```python
df = pd.read_excel('./data/iris.xlsx')
X = df.drop(columns=['clase'])
X.head()
```

```python
X.info()
```

## Clustering with K-means
### K optimization

```python
k_list = range(1, 10)
intertia_list = []

for k in k_list:
    model = KMeans(n_clusters=k, random_state=0)
    model.fit(X)
    intertia_list.append(model.inertia_)

plt.plot(k_list, intertia_list, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.xticks(k_list)
plt.show()
```

Best K will be 3

```python
model = KMeans(n_clusters=3, random_state=0)
model.fit(X)
```

```python
plt.scatter(X['long_sepalo'], X['anch_sepalo'], c=model.labels_, cmap='viridis')
plt.show()
```
